### Line chart

    * date/numeric column

```python
exp = [0,2,4,6,8,10]
sal = [10000,20000,25000,12000,35000,45000]
plt.plot(exp, sal)
plt.show()
```

```python
plt.plot(exp, sal)
plt.yticks([5000,10000,15000,20000,25000,30000,35000,40000,45000,50000], c='darkred')
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(exp, sal, c='darkblue')
plt.xticks(range(0,11), c='brown')
plt.yticks(range(5000,55000,5000), c='darkred')
plt.xlabel("exp",c='orange')
plt.ylabel("sal",c='darkorange')
plt.title("line chart", c='g')
plt.show()
```

```python
exp=[0,2,4,6,8]
sal=[10000,15000,18000,8000,35000]
plt.figure(figsize=(5,4))
plt.plot(exp,sal,c="darkgreen",linestyle="-",linewidth=2,marker="h",markersize=10,markerfacecolor="r",markeredgecolor="k")
plt.xticks(range(0,10),size=12)
plt.yticks(range(5000,50000,5000),size=12)
plt.xlabel("Exp",size=15,c="red")
plt.ylabel("Sal",size=15,c="brown")
plt.title("Sal Trend Over Exp",c="#0000FF",size=13)
plt.show()
```

```python
exp=[0,2,4,6,8]
sal=[10000,15000,18000,8000,35000]
sal2=[5000,12000,22000,22000,40000]

plt.figure(figsize=(5,4))
plt.plot(exp,sal,c="darkgreen",linestyle="-",linewidth=2,marker="h",markersize=7,markerfacecolor="r",markeredgecolor="k")
plt.plot(exp,sal2,c="b",linestyle="-",linewidth=2,marker="h",markersize=7,markerfacecolor="r",markeredgecolor="k")
plt.xticks(range(0,10),size=12)
plt.yticks(range(5000,50000,5000),size=12)
plt.xlabel("Exp",size=15,c="red")
plt.ylabel("Sal",size=15,c="brown")
plt.title("Sal Trend Over Exp",c="#0000FF",size=13)
plt.legend(["Me","Friend"],loc="upper left")
#plt.grid()
#plt.text(0,40000,"hi")

for x,y in zip(exp,sal):
    plt.text(x,y+1000,y)

for a,b in zip(exp,sal2):
    plt.text(a,b,b)
#plt.show()
plt.savefig("E:/e drive data/Download/Sal Trend Over Exp.jpeg")
```

#### Bar Chart
    * comparision of a categorical column based on aggragate value on same/other column:
                          aggragate- sum,min,max,count,avg,median,mode

```python
import pandas as pd
```

```python
df=pd.read_csv("e:/dataset/dataset/analysis/restaurant.csv")
df
```

```python
df.day.unique()
```

```python
s=df.day.value_counts()
```

```python
s
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,3))
#plt.bar(x=s.index,height=s.values,edgecolor="k",color="g")
plt.bar(x=s.index,height=s.values,edgecolor="k",color=["b","r","g","c"])
for a,b in zip(s.index,s.values):
    plt.text(a,b,b)
plt.show()
```

```python
plt.figure(figsize=(5,3))
A=df.groupby(by="day").total_bill.sum()
A.plot.bar(edgecolor="k",width=.5)
#for x,y in zip(A.index,A.values):
 #   plt.text(x,y,y)
```

```python
a=df.groupby(by="day").total_bill.sum()
a
```

```python
plt.figure(figsize=(5,3))
plt.bar(x=a.index,height=a.values,edgecolor="k",color=["b","g","r","pink"],align="edge",width=0.8)
for x,y in zip(a.index,a.values):
    plt.text(x,y+2,y)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.pie(x=pc_frq.values,labels=pc_frq.index,autopct="%.2f",startangle=162,wedgeprops={"edgecolor":"k"},colors=["r","b","g"]) #by default startangle is 0 
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.pie(x=pc_frq.values,labels=pc_frq.index,autopct="%.2f",explode=[0,0.1,0],startangle=162,wedgeprops={"edgecolor":"k"},colors=["r","b","g"]) #by default startangle is 0 
plt.show()
```
