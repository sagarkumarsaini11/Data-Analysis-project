## Data Encoding
    - categorical data convert into numeric data

#### Categorical data
    * nominal- we don't know order, ex-[red, green, white, yellow,pink]
    * ordinal  we know order, ex-[small,medium,large,xl]

```python
import pandas as pd
df=pd.read_csv("e:/dataset/dataset/analysis/clothes.txt")
df
```

## Encoding
    * label encoding
    * one hot coding

```python
df["size"].unique()
```

```python
df["size"]=df["size"].apply(str.lower)
df.brand=df.brand.apply(str.lower)
```

```python
df["size"].unique()
```

```python
df
```

```python
#label encoding
df["size"]=df["size"].map({"small":0,"medium":1,"large":2})
```

```python
df.brand=df.brand.map({"cantabil":0,"cob":1,"monte":2,"peter":3})
```

```python
df
```
