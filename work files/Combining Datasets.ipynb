## Combining Datasets
    * concat()

```python
import pandas as pd
```

```python
df_cs=pd.read_csv("e:/dataset/dataset/analysis/cs.txt")
df_cs
```

```python
df_civil=pd.read_csv("e:/dataset/dataset/analysis/civil.txt")
df_civil
```

```python
df_ec=pd.read_csv("e:/dataset/dataset/analysis/ec.txt")
df_ec
```

```python
pd.concat([df_cs,df_civil,df_ec],axis=0)
```

```python
pd.concat([df_cs,df_civil,df_ec],axis=0,ignore_index=True)
```

```python
pd.concat([df_cs,df_civil,df_ec],axis=0,ignore_index=True,sort=True)
```

```python
df_a=pd.DataFrame({"name":["monu", "sonu"],"age":[22,23]})
df_a
```
